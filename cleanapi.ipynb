```r
rm(list = ls())
source("tests.r")
source("simdatagen.r")
library(bnlearn)
library(caret)
library(pROC)
res1000 <- sim(1000,0.05,0.1)
res500 <- sim(500,0.05,0.1)
res200 <- sim(200,0.05,0.1)
res100 <- sim(200,0.05,0.1)
res50 <- sim(50,0.05,0.1)
res20 <- sim(20,0.05,0.1)
siz <- c(1000,500,200,100,50,20)
```

```r
acc<- rbind(res1000$accuracy, res500$accuracy, res200$accuracy, res100$accuracy, res50$accuracy, res20$accuracy)
```

```r
spe<- rbind(res1000$specificity, res500$specificity, res200$specificity, res100$specificity, res50$specificity, res20$specificity)
sen <- rbind(res1000$sensitivity, res500$sensitivity, res200$sensitivity, res100$sensitivity, res50$sensitivity, res20$sensitivity)
auc <- rbind(res1000$auc, res500$auc, res200$auc, res100$auc, res50$auc, res20$auc)
```

```r
write.csv(acc, "acc_at0.05nd1.csv")
write.csv(spe, "spe_at0.05nd1.csv")
write.csv(sen, "sen_at0.05nd1.csv")
write.csv(auc, "auc_at0.05nd1.csv")
```

```r
plot(siz, acc[,1], type="l", ylim=c(0,1),lwd=2.5, xlab="Size", ylab="Accuracy" ,main="Accuracy")
lines(siz, acc[,2], type="l", ylim=c(0,1),lwd=2.5, col = "red")
lines(siz, acc[,3], type="l", ylim=c(0,1),lwd=2.5, col = "blue")
lines(siz, acc[,4], type="l", ylim=c(0,1),lwd=2.5, col = "green")

plot(siz, spe[,1], type="l", ylim=c(0,1),lwd=2.5, xlab="Size", ylab="Specificity",main="Specificity")
lines(siz, spe[,2], type="l", ylim=c(0,1),lwd=2.5, col = "red")
lines(siz, spe[,3], type="l", ylim=c(0,1),lwd=2.5, col = "blue")
lines(siz, spe[,4], type="l", ylim=c(0,1),lwd=2.5, col = "green")

plot(siz, sen[,1], type="l", ylim=c(0,1),lwd=2.5, xlab="Size", ylab="Sensitivity",main="Sensitivity")
lines(siz, sen[,2], type="l", ylim=c(0,1),lwd=2.5, col = "red")
lines(siz, sen[,3], type="l", ylim=c(0,1),lwd=2.5, col = "blue")
lines(siz, sen[,4], type="l", ylim=c(0,1),lwd=2.5, col = "green")
```
